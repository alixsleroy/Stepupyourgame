```python
import numpy as np
def U(x,y):
    res = 100*y*y/(1+10*np.power(x,4))+0.001*np.power((x*x-9),2)
    return res
tau=0.1

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

x = np.arange(-1.5, 1.5, 0.1)
y = np.arange(-1, 1, 0.1)

X, Y = np.meshgrid(x, y)
Z = np.exp(-U(X,Y)*tau)

fig = plt.figure()
ax = Axes3D(fig,auto_add_to_figure=False)
fig.add_axes(ax)
ax.contourf(X, Y, Z, 100)
# ax.plot_surface(X, Y, Z)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def Up(x,y):
    Upx = -4000*y*y*x*x*x*np.power((1+10*np.power(x,4)),-2)+0.004*(x*x-9)*x
    Upy = 200*y/(1+10*np.power(y,4))

    return Upx,Upy

x=np.linspace(-10,10,1000)
y=np.linspace(-10,10,1000)
Upx,Upy = Up(x,y)
```

```python
plt.plot(x,Upx)
plt.plot(y,Upy)
```

```python
def marginal(x,beta): 
    rho = np.sqrt(10*np.power(x,4)+1)*np.exp(-beta*0.001*(x*x-9)*(x*x-9))
    return rho

x = np.linspace(-10,10,100)
plt.plot(x,marginal(x,1))
plt.title(" $\\rho_{\\beta}(x)$")
```
